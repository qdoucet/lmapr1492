```python
from pymatgen.ext.matproj import MPRester
with MPRester("VK9U10BEjITOk1kneKKl") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-924128")
    
structure.to(filename="HfNiSn.cif")
```

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
xrd = XRDCalculator(1.54060)
pattern = xrd.get_pattern(structure)
```

```python
print("Premier pic :\n")
print("Angle 2\u0398 = ", pattern.x[0],"°")
print("Intensité = ", pattern.y[0],"%")
print("Indices hkl = ", pattern.hkls[0])
print("d_hkl = ", pattern.d_hkls[0])
print("---------------")

print("Deuxième pic :\n")
print("Angle 2\u0398 = ", pattern.x[1],"°")
print("Intensité = ", pattern.y[1],"%")
print("Indices hkl = ", pattern.hkls[1])
print("d_hkl = ", pattern.d_hkls[1])
print("---------------")

print("Troisième pic :\n")
print("Angle 2\u0398 = ", pattern.x[2],"°")
print("Intensité = ", pattern.y[2],"%")
print("Indices hkl = ", pattern.hkls[2])
print("d_hkl = ", pattern.d_hkls[2])
print("---------------")
```
